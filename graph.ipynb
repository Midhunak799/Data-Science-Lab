```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([1,4,2,6,2,4,2,6])
plt.plot(x,y,marker='*',color='m',linestyle='-.',mec='g',mfc='g')
plt.title('Log',loc='center')
plt.xlabel('x-axis',color='g')
plt.ylabel('y-axis',color='b')
plt.show()

```
